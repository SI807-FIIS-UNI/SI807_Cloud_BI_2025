```python
# Nuevo cell (índice 0) - leer y mostrar gender_submission.csv
import pandas as pd
from IPython.display import display
from datetime import datetime

path_1 = r'../data/raw/gender_submission.csv'
path_2 = r'../data/raw/train.csv'
path_3 = r'../data/raw/test.csv' 

# Cargar
df_gender_submission = pd.read_csv(path_1)
df_titanic = pd.read_csv(path_2)
df_test = pd.read_csv(path_3)

# Mostrar contenido completo si no es muy grande, si no mostrar primeras 200 filas
def display_dataframe(df):
    if len(df) <= 5000:
        display(df)
    else:
        print("Archivo muy grande — mostrando primeras 200 filas:")
        display(df.head(200))
print("Muerte por género:")
display_dataframe(df_gender_submission)
print("Datos de prueba:")
display_dataframe(df_test)
print("Datos de entrenamiento:")
display_dataframe(df_titanic)

# Imprimir momento de ejecución

print("Ejecución del notebook:", datetime.now().strftime("%Y-%m-%d %H:%M:%S"))

```

```python
# Uniendo a partir de la columna "PassengerId"
merged_df = df_test.merge(df_gender_submission, on="PassengerId", how="inner", suffixes=("", "_gender"))
display_dataframe(merged_df)
```

```python
df_cabina = merged_df.groupby('Cabin')['Survived'].sum().reset_index(name='Conteo_Supervivientes')
display_dataframe(df_cabina)

df_class = merged_df.groupby('Pclass')['Survived'].sum().reset_index(name='Conteo_Supervivientes')
display_dataframe(df_class)
# Graficando con matplotlib
import matplotlib.pyplot as plt
# Gráfico de barras para supervivientes por clase
plt.figure(figsize=(8, 6))
plt.bar(df_class['Pclass'], df_class['Conteo_Supervivientes'], color='skyblue')
plt.xlabel('Clase de Pasajero')
plt.ylabel('Conteo de Supervivientes')
plt.title('Supervivientes por Clase de Pasajero')
plt.xticks(rotation=0)
plt.show()
```
