```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../Data/Raw/netflix_titles.csv")
df.head()
```

```python
missing = df.isna().sum().sort_values(ascending=False)
(missing / len(df) * 100).round(2)
```

## Gráficos

```python
sns.countplot(data=df, x="type")
plt.title("Cantidad por tipo (Movie vs TV Show)")
plt.show()
```

```python
sns.histplot(df["release_year"], bins=30)
plt.title("Distribución de años de lanzamiento")
plt.show()
```

## Anáñisis de duración

```python
def parse_duration(x):
    if pd.isna(x):
        return np.nan
    try:
        return int(x.split()[0])
    except:
        return np.nan


df["duration_minutes"] = df["duration"].apply(parse_duration)
sns.histplot(df["duration_minutes"].dropna(), bins=30)
plt.title("Duración (solo películas)")
plt.show()
```

## Correlación entre variables numéricas

```python
df_num = df.select_dtypes(include=[np.number])
sns.heatmap(df_num.corr(), annot=True, cmap="coolwarm", center=0)
plt.title("Matriz de correlación (numéricas)")
plt.show()
```

# Genero con mas frecuencia en la catálogo de Netflix

```python
# Separar los géneros que están separados por comas 

genre_series = df['listed_in'].str.split(',').explode().str.strip() 

# Contar los géneros más frecuentes 

top_genres = genre_series.value_counts().head(15) 

# Visualizar 

plt.figure(figsize=(10,6)) 
sns.barplot(y=top_genres.index, x=top_genres.values, palette="viridis") 
plt.title("🎭 Géneros más frecuentes en el catálogo de Netflix") 
plt.xlabel("Número de títulos") 
plt.ylabel("Género") 
plt.show()
```

# Conteo de países con más títulos

```python
# 📊 Conteo de países con más títulos 
country_series = df['country'].dropna().str.split(',').explode().str.strip() 
top_countries = country_series.value_counts().head(10) 

plt.figure(figsize=(10,6)) 
sns.barplot(y=top_countries.index, x=top_countries.values, palette="rocket") 
plt.title("🌍 Países con mayor número de títulos en Netflix") 
plt.xlabel("Número de títulos") 
plt.ylabel("País") 
plt.show() 
```

# Clasificaciones por edad

```python
# 📊 Clasificaciones por edad 
rating_counts = df['rating'].value_counts().head(10) 
plt.figure(figsize=(8,5)) 
sns.barplot(x=rating_counts.values, y=rating_counts.index, palette="crest") 
plt.title("🔖 Distribución por clasificación de edad (Rating)") 
plt.xlabel("Número de títulos") 
plt.ylabel("Clasificación") 
plt.show()
```
