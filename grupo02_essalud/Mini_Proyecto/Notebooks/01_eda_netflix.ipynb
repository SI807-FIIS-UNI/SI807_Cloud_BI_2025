```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/Mini_Proyecto/Data/Raw/netflix_titles.csv")
df.head()
```

## Gr√°ficos

```python
sns.countplot(data=df, x="type")
plt.title("Cantidad por tipo (Movie vs TV Show)")
plt.show()
```

```python
sns.histplot(df["release_year"], bins=30)
plt.title("Distribuci√≥n de a√±os de lanzamiento")
plt.show()
```

## An√°√±isis de duraci√≥n

```python
def parse_duration(x):
    if pd.isna(x):
        return np.nan
    try:
        return int(x.split()[0])
    except:
        return np.nan


df["duration_minutes"] = df["duration"].apply(parse_duration)
sns.histplot(df["duration_minutes"].dropna(), bins=30)
plt.title("Duraci√≥n (solo pel√≠culas)")
plt.show()
```

## Correlaci√≥n entre variables num√©ricas

```python
df_num = df.select_dtypes(include=[np.number])
sns.heatmap(df_num.corr(), annot=True, cmap="coolwarm", center=0)
plt.title("Matriz de correlaci√≥n (num√©ricas)")
plt.show()
```
