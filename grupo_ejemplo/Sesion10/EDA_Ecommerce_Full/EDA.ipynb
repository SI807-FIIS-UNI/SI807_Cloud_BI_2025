```python
path = '/content/drive/MyDrive/Ecommerce_Sales.csv'
df = pd.read_csv(path, encoding='latin1')
print('âœ… Datos cargados correctamente')
resumen_general(df)
```

```python
df.columns = [c.strip().replace(' ', '_').lower() for c in df.columns]
df.drop_duplicates(inplace=True)
plot_missing(df)
distribucion_numericas(df, ['sales', 'profit', 'quantity', 'discount'])
plot_correlaciones(df)
analisis_ventas_profit(df)
```

```python
cat_summary = df.groupby('category')[['sales','profit']].sum().sort_values(by='sales', ascending=False)
display(cat_summary)
sns.barplot(data=cat_summary.reset_index(), x='category', y='sales', palette='viridis')
plt.title('Ventas Totales por CategorÃ­a')
plt.show()
seg = df.groupby('segment')[['sales','profit']].sum().sort_values(by='profit', ascending=False)
sns.barplot(x=seg.index, y=seg['profit'], palette='crest')
plt.title('Rentabilidad por Segmento de Cliente')
plt.show()
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])
monthly = df.groupby(df['order_date'].dt.to_period('M'))[['sales','profit']].sum()
monthly.index = monthly.index.to_timestamp()
fig, ax1 = plt.subplots()
ax1.plot(monthly.index, monthly['sales'], label='Ventas', color='blue')
ax2 = ax1.twinx()
ax2.plot(monthly.index, monthly['profit'], label='Ganancia', color='red')
plt.title('EvoluciÃ³n mensual de Ventas y Ganancia')
plt.show()
top_categorias(df, 'category', 'profit')
top_categorias(df, 'sub_category', 'profit')
clean_path = '/content/drive/MyDrive/ecommerce_clean.csv'
df.to_csv(clean_path, index=False)
print(f'âœ… Archivo limpio exportado a: {clean_path}')
```
