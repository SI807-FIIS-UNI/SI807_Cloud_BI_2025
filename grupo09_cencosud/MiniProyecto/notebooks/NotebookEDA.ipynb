```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar dataset (ajusta el nombre del archivo)
ruta_csv = "/content/sales_data_sample.csv"   # ← reemplaza con tu nombre real
df = pd.read_csv(ruta_csv,encoding='latin-1')
df.head()
```

```python
# Resumen general
df.info()

# Valores nulos
df.isna().sum()

# Tipos de datos
df.dtypes

# Descripción estadística
df.describe(include='all').transpose()
```

```python
num_cols = df.select_dtypes(include=np.number).columns
cat_cols = df.select_dtypes(include='object').columns

print("🔢 Variables numéricas:", list(num_cols))
print("🔠 Variables categóricas:", list(cat_cols))
```

```python
# Histogramas numéricos
df[num_cols].hist(figsize=(12, 10), bins=20, edgecolor='black')
plt.suptitle("Distribución de variables numéricas")
plt.show()

# Correlación
plt.figure(figsize=(8,6))
sns.heatmap(df[num_cols].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matriz de correlación")
plt.show()
```

```python
for col in cat_cols:
    print(f"\n📊 {col}")
    print(df[col].value_counts().head(5))
    sns.countplot(data=df, y=col, order=df[col].value_counts().index[:5])
    plt.title(f"Distribución de {col}")
    plt.show()
```

```python
# Eliminar columnas donde más del 50% de los valores son nulos
umbral = 0.5
df_clean = df.dropna(axis=1, thresh=len(df) * (1 - umbral))

print(f"✅ Columnas restantes: {df_clean.shape[1]}")
```

```python
df_clean.to_csv("sales_data_sample_clean.csv", index=False, encoding='utf-8')
print("💾 Archivo limpio guardado como sales_data_sample_clean.csv")
```
