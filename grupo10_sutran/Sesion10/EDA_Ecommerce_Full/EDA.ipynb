```python
path = 'C:\\Users\\jairo\\Documents\\Proyectos\\aradiel_25-2\\taller_2\\archive\\Amazon Sale Report.csv'
df = pd.read_csv(path, encoding='latin1')
print('‚úÖ Datos cargados correctamente')
resumen_general(df)
```

```python
# Normalizar nombres de columnas (sin espacios ni may√∫sculas)
df.columns = [c.strip().replace(' ', '_').lower() for c in df.columns]

# Eliminar duplicados
df.drop_duplicates(inplace=True)

# Mostrar valores faltantes
plot_missing(df)

# Revisar la distribuci√≥n de variables num√©ricas relevantes
# En tu CSV, las columnas num√©ricas principales son: qty, amount
distribucion_numericas(df, ['qty', 'amount'])

# Mapa de correlaciones (entre qty y amount)
plot_correlaciones(df)

# An√°lisis general de ventas
analisis_ventas(df)

# Top 10 categor√≠as con m√°s monto vendido
top_categorias(df, group_field='category', metric='amount', n=10)
```

```python
# --- An√°lisis de Ventas por Categor√≠a ---
cat_summary = (
    df.groupby('category')[['amount', 'qty']]
    .sum()
    .sort_values(by='amount', ascending=False)
)

display(cat_summary)

sns.barplot(
    data=cat_summary.reset_index(),
    x='category',
    y='amount',
    hue='category',
    palette='viridis',
    legend=False
)
plt.title('üí∞ Ventas Totales por Categor√≠a')
plt.xlabel('Categor√≠a')
plt.ylabel('Monto Total (Amount)')
plt.xticks(rotation=45, ha='right')
plt.show()


# --- An√°lisis de Rentabilidad / Comparativo por Tipo de Cliente ---
# En este dataset no existe "segment", as√≠ que usaremos 'b2b' o 'sales_channel' como alternativa
if 'b2b' in df.columns:
    seg_summary = (
        df.groupby('b2b')[['amount', 'qty']]
        .sum()
        .sort_values(by='amount', ascending=False)
    )
    x_col = 'b2b'
    title = 'üíº Ventas por Tipo de Cliente (B2B vs B2C)'
elif 'sales_channel' in df.columns:
    seg_summary = (
        df.groupby('sales_channel')[['amount', 'qty']]
        .sum()
        .sort_values(by='amount', ascending=False)
    )
    x_col = 'sales_channel'
    title = 'üì¶ Ventas por Canal de Venta'
else:
    print("‚ö†Ô∏è No se encontr√≥ columna 'b2b' ni 'sales_channel' para agrupar.")
    seg_summary = None

if seg_summary is not None:
    sns.barplot(
        data=seg_summary.reset_index(),
        x=x_col,
        y='amount',
        hue=x_col,
        palette='crest',
        legend=False
    )
    plt.title(title)
    plt.xlabel(x_col.replace('_', ' ').title())
    plt.ylabel('Monto Total (Amount)')
    plt.xticks(rotation=45, ha='right')
    plt.show()
```

```python
# --- Conversi√≥n de fechas ---
# Convertir la columna 'date' a formato datetime
df['date'] = pd.to_datetime(df['date'], errors='coerce')

# --- Agrupaci√≥n mensual ---
monthly = (
    df.groupby(df['date'].dt.to_period('M'))[['amount', 'qty']]
    .sum()
)
monthly.index = monthly.index.to_timestamp()

# --- Gr√°fico: evoluci√≥n mensual de ventas y cantidad ---
fig, ax1 = plt.subplots(figsize=(10,5))

ax1.plot(monthly.index, monthly['amount'], label='Monto total (ventas)', color='blue', marker='o')
ax1.set_xlabel('Mes')
ax1.set_ylabel('Ventas (Amount)', color='blue')

ax2 = ax1.twinx()
ax2.plot(monthly.index, monthly['qty'], label='Cantidad vendida', color='orange', marker='o')
ax2.set_ylabel('Cantidad de productos', color='orange')

plt.title('üìà Evoluci√≥n Mensual de Ventas y Cantidad Vendida')
fig.tight_layout()
plt.show()

# --- Top categor√≠as por ventas ---
top_categorias(df, 'category', 'amount')

# --- Exportar dataset limpio ---
clean_path = r'C:\Users\jairo\Documents\Proyectos\aradiel_25-2\taller_2\archive\ecommerce_clean.csv'
df.to_csv(clean_path, index=False, encoding='utf-8-sig')
print(f'‚úÖ Archivo limpio exportado a: {clean_path}')
```
